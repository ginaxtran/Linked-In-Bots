```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('linked_in_data.csv')

df['Connections'] = pd.to_numeric(df['Connections'], errors='coerce')
df['Followers'] = pd.to_numeric(df['Followers'], errors='coerce')

# Plotting the distribution of 'Connections'
plt.figure(figsize=(10, 6))
sns.histplot(df['Connections'], kde=True, color='blue')
plt.title('Distribution of Connections')
plt.xlabel('Connections')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10, 6))
sns.histplot(df['Followers'], kde=True, color='green')
plt.title('Distribution of Followers')
plt.xlabel('Followers')
plt.ylabel('Frequency')
plt.show()

# Creating a correlation matrix
correlation_matrix = df[['Connections', 'Followers']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix between Connections and Followers')
plt.show()
```
