```python
import pandas as pd
import random
from faker import Faker
fake = Faker()

def create_profile():
    full_name = fake.name()
    industry = random.choice(['Technology', 'Finance', 'Healthcare', 'Education', 'Engineering'])
    workplace = f"{fake.company()} {industry}"
    location = f"{fake.city()}, {fake.state_abbr()}, {fake.country()}"
    connections = random.randint(300, 1000)  # More realistic range for professional profiles
    photo = "Yes"  # Assuming legitimate profiles are more likely to have a photo
    followers = random.randint(500, 10000)  # Higher followers count for legitimate profiles
    about = fake.paragraph(nb_sentences=3)
    role = random.choice(['Manager', 'Engineer', 'Analyst', 'Consultant', 'Specialist'])
    experiences = {
        '0': {
            'Role': f'Senior {role}',
            'Workplace': workplace,
            'Duration': fake.date_between(start_date='-5y', end_date='today').strftime('%b %Y') + " - Present · " + str(random.randint(3, 10)) + ' yrs',
            'Workplace Location': location,
            'Description': fake.text(max_nb_chars=150)
        },
        '1': {
            'Role': role,
            'Workplace': fake.company() + " " + industry,
            'Duration': fake.date_between(start_date='-10y', end_date='-5y').strftime('%b %Y') + " - " + fake.date_between(start_date='-5y', end_date='today').strftime('%b %Y') + ' · ' + str(random.randint(2, 5)) + ' yrs',
            'Workplace Location': fake.city() + ", " + fake.state_abbr() + ", " + fake.country(),
            'Description': fake.text(max_nb_chars=150)
        }
    }
    num_experiences = len(experiences)
    education = {
        '0': {
            'Institute': f"{fake.company()} University",
            'Degree': 'Bachelor of ' + fake.word(ext_word_list=['Science', 'Arts', 'Engineering', 'Business']),
            'Duration': str(random.randint(1995, 2005)) + " - " + str(random.randint(2005, 2009)),
            'Description': fake.sentence()
        }
    }
    num_education = len(education)
    skills = {f'Skill {i}': fake.job() for i in range(random.randint(3, 5))}
    num_skills = len(skills)
    interests = {f'{i}': {'Interest': fake.company(), 'ID': str(fake.random_int(min=100000, max=999999)), 'Category': 'industry', 'URL': 'https://www.linkedin.com/company/' + fake.domain_name()} for i in range(random.randint(2, 4))}
    num_interests = len(interests)
    label = 1

    return {
        'Intro': str({'Full Name': full_name, 'Workplace': workplace, 'Location': location, 'Connections': f"{connections}+", 'Photo': photo, 'Followers': followers}),
        'Full Name': full_name,
        'Workplace': workplace,
        'Location': location,
        'Connections': connections,
        'Photo': photo,
        'Followers': followers,
        'About': about,
        'Experiences': str(experiences),
        'Number of Experiences': num_experiences,
        'Educations': str(education),
        'Number of Educations': num_education,
        'Skills': str(skills),
        'Number of Skills': num_skills,
        'Interests': str(interests),
        'Number of Interests': num_interests,
        'Label': label
    }

profiles = [create_profile() for _ in range(2000)]
df = pd.DataFrame(profiles)

df.to_csv('CLP_linkedin_profiles.csv', index=False)
```
